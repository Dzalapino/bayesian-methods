```python
import matplotlib.pyplot as plt

class BayesianReasoning:
    def __init__(self, p_cause, p_effect, p_effect_given_cause = None, p_cause_given_effect = None):
        self.p_cause = p_cause
        self.p_effect = p_effect
        self.p_effect_given_cause = p_effect_given_cause
        self.p_cause_given_effect = p_cause_given_effect
        self.update_probability()

    def update_probability(self):
        # Applying Bayes' theorem to update the probabilities
        if (self.p_effect_given_cause is None):
          self.p_effect_given_cause = self.p_cause_given_effect * self.p_effect / self.p_cause
          self.plot_probability('P(E|C)')
          print(f'Calculated probability P(E|C) = {self.p_effect_given_cause}')
        else:
          self.p_cause_given_effect = self.p_effect_given_cause * self.p_cause / self.p_effect
          self.plot_probability('P(C|E)')
          print(f'Calculated probability P(C|E) = {self.p_cause_given_effect}')

    def plot_probability(self, title):
        plt.plot([0, 1], [self.p_cause, self.p_cause], label='Prior probability of cause(p(C))')
        plt.plot([0, 1], [self.p_effect, self.p_effect], label='Prior probability of effect(p(E))')
        plt.plot([0, 1], [self.p_effect_given_cause, self.p_effect_given_cause], label='Conditional probability (p(E|C))')
        plt.plot([0, 1], [self.p_cause_given_effect, self.p_cause_given_effect], label='Conditional probability (p(C|E))')
        plt.title('Bayesian Reasoning after updating ' + title)
        plt.xlabel('Events')
        plt.ylabel('Probability')
        plt.legend()
        plt.show()
```
