```python
import numpy as np
import pymc as pm
import arviz as az

zeros = np.zeros(30)
ones = np.ones(70)
occurences = np.concatenate((zeros, ones))

with pm.Model() as model:
  probability = pm.Uniform('probability', lower=0, upper=1)
  observed = pm.Bernoulli('observed', probability, observed=occurences)
  idata = pm.sample(2000, tune=2500)
  az.plot_trace(idata, show=True)
```
